```python
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage as ndi 
from skimage.data import data_dir
from skimage import feature
from skimage import io


im = io.imread(os.path.join(data_dir, "shawnMendes.png"), as_grey=True)
# Compute the Canny filter for two values of sigma
edges1 = feature.canny(im)
edges2 = feature.canny(im, sigma=3)

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(20, 5), sharex=True, sharey=True)
ax1.imshow(im, cmap=plt.cm.gray)
ax1.axis('off')
ax1.set_title('Original', fontsize=20)

ax2.imshow(edges1, cmap=plt.cm.gray)
ax2.axis('off')
ax2.set_title('Canny filter, $\sigma=1$', fontsize=20)

ax3.imshow(edges2, cmap=plt.cm.gray)
ax3.axis('off')
ax3.set_title('Canny filter, $\sigma=3$', fontsize=20)

fig, (ax4, ax5) = plt.subplots(nrows=1, ncols=2, figsize=(15, 10), sharex=True, sharey=True)

ax4.imshow(im, cmap=plt.cm.gray)
ax4.axis('off')
ax4.set_title('Original', fontsize=20)

sx = ndi.sobel(im, axis = 0, mode = 'constant') 
sy = ndi.sobel(im, axis = 1, mode = 'constant') 
sob = np.hypot(sy,sx)
ax5.imshow(sob,cmap='gray')
ax5.axis('off')
ax5.set_title('Sobel')

plt.show()
```
