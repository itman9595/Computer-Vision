```python
from scipy import ndimage 
import numpy as np 
import matplotlib.image as img 
import matplotlib.pyplot as plt 
from skimage import color 
from skimage import io 
from skimage import util 

import matplotlib.image as pltimg 
image = io.imread('shawnMendes.jpg', as_grey=True) 
image = util.random_noise(image, mode='s&p', seed=None, clip=True) 



#Filter A 
A = np.array([[1.*1/6,2.*1/6,1.*1/6],[2.*1/6,4.*1/6,2.*1/6],[1.*1/6,2.*1/6,1.*1/6]]) 
flt_1 = ndimage.convolve(image, A, mode='nearest', cval=0.0) 


#Filter B 
B = np.array([[0.*1/5,1.*1/5,0.*1/5],[1.*1/5,1.*1/5,1.*1/5],[0.*1/5,1.*1/5,0.*1/5]]) 
flt_2 = ndimage.convolve(image, B, mode='nearest', cval=0.0) 

#Filter C 
flt_3 = ndimage.median_filter(image, 3) 

#Filter D 
C = np.array([[-1.,-2.,-1.],[-2.,12.,-2.],[-1.,-2.,-1.]]) 
flt_4 = ndimage.convolve(image, C, mode='nearest', cval=0.0) 

#Filter E 
D = np.array([[-1.,-2.,-1.],[0.,0.,0.],[1.,2.,1.]]) 
flt_5 = ndimage.convolve(image, D, mode='nearest', cval=0.0) 




fig = plt.figure(figsize=(35,35)) 
plt.subplot(711) 
plt.title('Original') 
plt.imshow(image,cmap='gray') 
plt.subplot(712) 
plt.title('Filter A') 
plt.imshow(flt_1,cmap='gray') 
plt.subplot(713) 
plt.title('Filter B') 
plt.imshow(flt_2,cmap='gray') 
plt.subplot(714) 
plt.title('Filter C') 
plt.imshow(flt_3,cmap='gray') 
plt.subplot(715) 
plt.title('Filter D') 
plt.imshow(flt_4,cmap='gray') 
plt.subplot(716) 
plt.title('Filter E') 
plt.imshow(flt_5,cmap='gray') 
plt.show()
```
